```python
import pandas as pd
college_df = pd.read_csv('/content/CollegeBasketballPlayers2009-2021.csv')
college_df.head()
college_df.info()

# Remove players with no draft pick (non-drafted players)
college_df = college_df[college_df['pick'].notnull()]
college_df.head()
college_df.info()
```

```python
import pandas as pd

# Read the Excel file with the specified engine
draft_2010 = pd.read_csv('/content/Merged file.csv')

# Define the new column names
new_column_names = ['Rk', 'Pk', 'Tm', 'Player', 'College', 'Yrs', 'G', 'MP', 'PTS', 'TRB', 'AST', 'FG%', '3P%', 'FT%', 'MP', 'PTS', 'TRB', 'AST', 'WS', 'WS/48', 'BPM', 'VORP']

# Assign the new column names to the DataFrame
draft_2010.columns = new_column_names
draft_2010 = draft_2010.drop(0, axis=0)

# Reset index after dropping the row
draft_2010 = draft_2010.reset_index(drop=True)


# Check the new column names
draft_2010 = draft_2010.dropna(subset=['College'])
draft_2010 = draft_2010.dropna(subset=['Yrs'])


# Check the updated DataFrame
draft_2010.head()
draft_2010.info()


# Check the first few rows
```

```python


def classify_success(row):
    pick = pd.to_numeric(str(row["Pk"]), errors='coerce')  # Draft pick number
    ppg = pd.to_numeric(str(row["PTS"]), errors='coerce')  # Points per game
    rpg = pd.to_numeric(str(row["TRB"]), errors='coerce')  # Rebounds per game
    apg = pd.to_numeric(str(row["AST"]), errors='coerce')  # Assists per game
    ws = pd.to_numeric(str(row["WS"]), errors='coerce')  # Win Shares

    if pick <= 5:
        return 1 if (ppg >= 15 or rpg >= 5 or apg >= 5 or ws >= 40) else 0
    elif 6 <= pick <= 15:
        return 1 if (ppg >= 10 or rpg >= 4 or apg >= 4 or ws >= 25) else 0
    elif 16 <= pick <= 30:
        return 1 if (ppg >= 8 or rpg >= 3 or apg >= 3 or ws >= 15) else 0
    else:
        return 1 if (ppg >= 5 or rpg >= 2 or apg >= 2 or ws >= 5) else 0

# Apply function to dataset
draft_2010["Success"] = draft_2010.apply(classify_success, axis=1)

# Print results
print(draft_2010[["Player","Pk", "PTS", "TRB", "AST", "WS", "Success"]])

#save as draft_2010.csv
draft_2010.to_csv('draft_2010.csv', index=False)
```

```python
college_df = college_df.rename(columns={'player_name': 'Player'})

# Merge college_df with draft_2010 on 'Player' column to add the Success column
college_df_merge = college_df.merge(draft_2010[['Player','Pk', 'Success']], on='Player', how='left')

# Check the first few rows of the updated college_df
#college_df.head()

# Check the information of the merged dataframe
college_df.info()
```

```python
college_df_merge = college_df_merge.dropna(subset=['Success'])
college_df_merge.fillna(0, inplace=True)
# Split dataset into different draft pick groups
top_5 = college_df_merge[college_df_merge["pick"] <= 5]
lottery = college_df_merge[(college_df_merge["pick"] > 5) & (college_df_merge["pick"] <= 15)]
first_round = college_df_merge[(college_df_merge["pick"] > 15) & (college_df_merge["pick"] <= 30)]
second_round = college_df_merge[college_df_merge["pick"] > 30]

```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Convert 'Pk' to numeric
#college_df["Pk"] = pd.to_numeric(college_df["Pk"], errors="coerce")

# Drop any rows where 'Pk' is NaN after conversion
#college_df = college_df.dropna(subset=["Pk"])

# Convert back to integer (after handling NaNs)
#college_df["Pk"] = college_df["Pk"].astype(int)

# Define draft pick groups
pick_groups = {
    "Top 5": college_df_merge[college_df_merge["pick"] <= 5],
    "Lottery (6-15)": college_df_merge[(college_df_merge["pick"] > 5) & (college_df_merge["pick"] <= 15)],
    "First Round (16-30)": college_df_merge[(college_df["pick"] > 15) & (college_df_merge["pick"] <= 30)],
    "Second Round (31+)": college_df_merge[college_df_merge["pick"] > 30]
}
models = {}


# Train and evaluate models for each group
for group_name, group_df in pick_groups.items():
    if group_df.shape[0] < 10:  # Skip small datasets to avoid overfitting
        continue

    X = group_df.drop(columns=["Success", "Player",'yr','ht','num', "Pk",'conf','team','type'])  # Drop non-feature columns
    y = group_df["Success"]
    print(X.shape)
    print(y.shape)

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.75, random_state=42)

    # Train XGBoost model
    model = xgb.XGBClassifier(use_label_encoder=False, eval_metric="logloss")
    model.fit(X_train, y_train)

    # Predict and evaluate accuracy
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)

    print(f"{group_name} Accuracy: {accuracy:.2f}")
    models[group_name] = model

```

```python

new_data_2017 = college_df[(college_df['year'] == 2017) & (college_df['pick'].notnull())]
# Function to assign a model based on draft pick
def get_pick_group(pk):
    if pk <= 5:
        return "Top 5"
    elif 6 <= pk <= 15:
        return "Lottery (6-15)"
    elif 16 <= pk <= 30:
        return "First Round (16-30)"
    else:
        return "Second Round (31+)"


# Make predictions for each player
for index, row in new_data_2017.iterrows():
    group_name = get_pick_group(row["pick"])

    if group_name in models:
        model = models[group_name]

        # Prepare row for prediction
        features = row.drop(labels=["Success", "Player",'yr','ht','num', "Pk",'conf','team','type','Unnamed: 64','Unnamed: 65'], errors="ignore").values.reshape(1, -1)


        # Predict success
        new_data_2017.at[index, "Success"] = model.predict(features)[0]

# Save updated dataset with "Success" predictions
new_data_2017.to_csv("new_data_with_success.csv", index=False)
```
